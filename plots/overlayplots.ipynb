```python
plt.figure(figsize=(20, 20))
np.random.seed(0)

# Define binedges
binedges = [(42, 61)]

# Define values for first model point
values = [1e5]

# Define secondbinedges
binedges2 = [(145, 245)]

# Define values for second model point
values2 = [10**2.5]
color='purple'
xmin=0
xmax=350
xstep=50
ymin=0
ymax=6
img = imread(im)

plt.imshow(img,zorder=0, extent=[0, 350, 0, 6], aspect='auto')
ytix = [i for i in range(6)]
print ytix
xtix = [50*i for i in range(8)]
plt.xlim=((0,350))
plt.ylim=((0,7))
#plt.yscale=('log')
plt.xticks(xtix, fontsize=18)
plt.yticks(ytix, fontsize=18)
for index in range(len(binedges)):
    x = binedges[index]
    y = values[index]
    plt.axhline(y = np.log10(y), xmin = (x[0] - xmin)/(xmax - xmin), xmax = (x[1] - xmin)/(xmax - xmin), color=color, linewidth=10)
for index in range(len(binedges2)):
    x = binedges2[index]
    y = values2[index]
    plt.axhline(y = np.log10(y), xmin = (x[0] - xmin)/(xmax - xmin), xmax = (x[1] - xmin)/(xmax - xmin), color=color, linewidth=10, linestyle='--')
#plt.axhline(y = np.log10(y), xmin = x[0], xmax = x[1], color='red', linewidth=10, label="MA5 stops (750, 1)")

#plt.legend(loc=(170/xmax, 3.5))
plt.title(r'CMS Preliminary L=35.9fb$^{-1}$ 13 TeV', fontsize=30)
plt.xlabel(r'$\mathrm{M}_{\mathrm{T2}}$ (ll) / GeV', fontsize=30)
plt.ylabel(r'$\mathrm{log}_{10}$(Number of events)', fontsize=30)
legendpos = (160, 3.5)
plt.axhline(y =legendpos[1], xmin=legendpos[0]/xmax, xmax=(legendpos[0] + xmax/10.)/xmax, color=color, linewidth=10)
plt.annotate(r'$\tilde{t}\to\tilde{\chi}_1^0$ (750, 1), MadAnalysis5 v1.6', (legendpos[0] + xmax/9., legendpos[1]), fontsize=26)
legendpos = (legendpos[0], legendpos[1] - ymax/20.)
plt.axhline(y =legendpos[1], xmin=legendpos[0]/xmax, xmax=(legendpos[0] + xmax/10.)/xmax, color=color, linestyle='--', linewidth=10)
plt.annotate(r'$\tilde{t}\to\tilde{\chi}_1^0$ (600, 300), MadAnalysis5 v1.6', (legendpos[0] + xmax/9., legendpos[1]), fontsize=26)

plt.show()
plt.savefig('ma5ontopofcms.png')
plt.savefig('ma5ontopofcms.pdf')
```
